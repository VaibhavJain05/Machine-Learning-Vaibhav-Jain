```python
train = train_set.to_numpy() #converting datasets from csv to numpy arrays
test = test_set.to_numpy()
```

```python
x=(train[:,1:])/255
xt=(test[:,1:])/255
```

```python
m=x.shape[0]
mt=xt.shape[0]
```

```python
y=train[:,0]
yt=test[:,0]
```

```python
#applying one vs all 
def one_vs_all(m):
  Z= np.zeros((m,26))
  for i in range(m):  
    Z[i][y[i]-1] = 1
    i+=1
  return Z
yp=one_vs_all(m)
ypt=one_vs_all(mt)
```

```python
def sigm_(a):
  return 1/(1+np.exp(-a))
```

```python
def sigm_derv(a):
  return a*(1-a)
```

```python
epsilon = np.random.random(1)[0]
```

```python
def nn(x,iterations,alpha,m,c):
  #gradient descent
  #paramaters
  theta1=np.random.rand(26,x.shape[1])*(2*epsilon) - epsilon
  bias1=np.random.rand(26,1)*(2*epsilon) - epsilon
  theta2=np.random.rand(26,26)*(2*epsilon) - epsilon
  bias2=np.random.rand(26,1)*(2*epsilon) - epsilon
  theta3=np.random.rand(26,26)*(2*epsilon) - epsilon
  bias3=np.random.rand(26,1)*(2*epsilon) - epsilon
  for i in range(iterations):
    #forward propagation
    Z2 = theta1@(x.T) + bias1
    A2 = sigm_(Z2)
    Z3 = theta2@(A2) + bias2
    A3 = sigm_(Z3)
    Z4 = theta3@(A3) + bias3
    h_x = sigm_(Z4)
    
    #back propagation
    dZ4 = h_x - c.T
    dtheta3 = dZ4@(A3.T)
    dbias3 = np.sum(dZ4)
    dZ3 = ((theta3.T)@(dZ4))*(sigm_derv(A3))
    dtheta2 = dZ3@(A2.T)
    dbias2 = np.sum(dZ3)
    dZ2 = ((theta2.T)@(dZ3))*(sigm_derv(A2))
    dtheta1 = dZ3@(x)
    dbias1  = np.sum(dZ2)
 

    #updating parameters
    theta1 -= (alpha/m)*(dtheta1)
    bias1 -= (alpha/m)*(dbias1)    
    theta2 -= (alpha/m)*(dtheta2)  
    bias2 -= (alpha/m)*(dbias2)  
    theta3 -= (alpha/m)*(dtheta3)
    bias3 -= (alpha/m)*(dbias3)
    
    cost=(-1/m)*(np.sum(c.T*(np.log(h_x)) + (1-c.T)*(np.log(1-h_x))))
    if i%10==0:
      print(cost)
  return theta1,bias1,theta2,bias2,theta3,bias3
```

```python
q1,b1,q2,b2,q3,b3= nn(xt,300,0.7,mt,ypt)
#forward propagation on test
Z2 = q1@(xt.T) + b1
A2 = sigm_(Z2)
Z3 = q2@(A2) + b2
A3 = sigm_(Z3)
Z4 = q3@(A3) + b3
y_pred= sigm_(Z4)
```

```python
y_pred
```

```python
Y_P=y_pred.T
```

```python
max_value= Y_P.max(axis=1)
for i in range(mt):
  for j in range(26):
    if Y_P[i][j]== max_value[i] :
      Y_P[i][j]=1
    else :
      Y_P[i][j]=0
print(Y_P)
```

```python
w = np.empty_like(yt)
for i in range(mt):
  for j in range(26):
    if Y_P[i][j]==1:
      w[i]= j+1
print(w)
```

```python
d= w-yt
g=0
for i in range(mt):
  if d[i]==0:
    g+=1
acc=(g/mt)*100
acc
```
