```python
def d2(A,B):  #degree 2 features

  return A*B
```

Defining a function for implementing standization

```python
def mean_norm(A):       #applying mean standardization and using standard deviation in place of range
  A_mean = A.mean(axis=0)
  A_std=A.std(axis=0)
  A_scale= (A-A_mean)/A_std
  return A_scale
```

Defining the function for calculating cost function and carrying out gradient descent

```python
def polynomial_regression(A,B,C,iterations,alpha):
   m=len(A)
   theta0 = np.zeros(A.shape[1])
   theta = np.vstack(theta0)
   for i in range(iterations) :
     h_x = np.dot(A, theta)
     error = h_x - B
     cost = (1/(2*m))*np.sum(np.square(error))
     grad= (1/m)*np.dot((A).T, error)
     theta = theta - (alpha)*(grad)
     print(cost)
   D = np.dot(C, theta)
   return D
   #A= input matrix
   #B= output matrix
   #C= testing matrix
```
