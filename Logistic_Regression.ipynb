```python
x = (train[:,1:])/255   #training sets
x.shape
```

```python
xt = (test[:,1:])/255  #testing sets
xt.shape
```

```python
a = np.vstack(np.ones(m))
X= np.append(a,x,axis=1)
```

```python
at = np.vstack(np.ones(mt))
Xt= np.append(at,xt,axis=1)
```

```python
y = train[:,0]  #training output
Y= np.vstack(y)
```

```python
yt= test[:,0]   #testing output
Yt = np.vstack(yt)
```

```python
Z= np.zeros((m,26)) #creating a zero matrix 
Z.shape
```

```python
for i in range(m):  #applying one vs all 
  Z[i][Y[i]-1] = 1
  i+=1
Z
```

gradient descent

```python
def lr(A,B,C,iterations,alpha):
  theta = np.zeros((A.shape[1], B.shape[1]))
  for i in range(iterations):
    z= A@theta
    h_x = 1/(1+ np.exp(-z))
    cost = (-1/m)*(np.sum(B*(np.log(h_x)) + (1-B)*(np.log(1-h_x))))
    grad = (1/m)*(np.dot(A.T,(h_x -B)))
    theta = theta - (alpha*grad)
    print(cost)
  D= 1/(1+ np.exp(-C@theta))
  return D
```

```python
Y_pred = lr(X,Z,Xt,1000,0.9)
Y_pred
```

```python
max_value= Y_pred.max(axis=1)
for i in range(mt):
  for j in range(26):
    if Y_pred[i][j]== max_value[i] :
      Y_pred[i][j]=1
    else :
      Y_pred[i][j]=0
print(Y_pred)
```

```python
w = np.empty_like(Yt)
```

```python
for i in range(mt):
  for j in range(26):
    if Y_pred[i][j]==1:
      w[i]= j+1
print(w)
```

```python
d=w-Yt
d
```

```python
g=0
for i in range(mt):
  if d[i]==0:
    g+=1
print(g)
```

```python
acc = (g/mt)*100
acc
```
