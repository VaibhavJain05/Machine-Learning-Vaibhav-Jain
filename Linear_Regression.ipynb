```python
def mean_scale(X):   #applying mean normalization and using standard deviation in place of range
  X_mean= X.mean(axis=0)
  X_std= X.std(axis=0)
  X_scale = (X-X_mean)/(X_std)
  return(X_scale)
```

```python
X_test= mean_scale(X1) #X1 is testing dataset
X_test.shape
```

```python
X_train=mean_scale(X) #X is training dataset
print(X_train,X_train.shape)
     
```

```python
P0 = np.ones(X_train.shape[0])
P= np.vstack(P0)
print(P,P.shape)
```

```python
Q0 = np.ones(X_test.shape[0])
Q= np.vstack(Q0)
```

```python
X_Train1 = np.append(P,X_train,axis=1)   #adding bias inputs to the arrays
print(X_Train1)
```

```python
X_Test1 = np.append(Q,X_test,axis=1)
X_Test1
```

Gradient decsent


```python
def linear_regression(A,B,C,iterations,alpha):
   m=len(A)
   theta0 = np.zeros(A.shape[1])
   theta = np.vstack(theta0)
   for i in range(iterations) :
     h_x = np.dot(A, theta)
     error = h_x - B
     cost = (1/(2*m))*np.sum(np.square(error))
     grad= (1/m)*np.dot((A).T, error)
     theta = theta - (alpha)*(grad)
     print(cost)
   D = np.dot(C, theta)
   return D
   #A= input matrix
   #B= output matrix
   #C= testing matrix
```

```python
LR=linear_regression( X_Train1,Y,X_Test1,3000,0.1 )
LR
```

```python
diff = LR-Y1
diff
```

```python
error_rms = (np.sum((diff)**2)/(2*len(Y1)))**(1/2)
error_rms 
```
