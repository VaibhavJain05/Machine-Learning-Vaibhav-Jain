```python
train_set= pd.read_csv(p1,header= None)
test_set= pd.read_csv(p2,header=None)
```

```python
train = train_set.to_numpy() #converting datasets from csv to numpy arrays
test = test_set.to_numpy()
```

```python
x= (train[:,1:])/255  #slicing and scaling the datasets
xt= (test[:,1:])/255
xt
```

```python
y=train[:,0]
yt=test[:,0]
```

```python
m= len(y)  #range of trainsets
mt = 100   #range of testsets
```

```python
def dist():  #defining a function for euclidian distance
  w=[]
  for i in range(mt):
    for j in range(m):
      diff=xt[i,:]-x[j,:]
      dis=np.sqrt(np.sum(np.square(diff)))
      w.append((i,dis,j))
  return w 
s=sorted(dist())  #arranging the distances according to ascending order
k=9
g=[]
for i in range(mt):  #getting first k nearest neighbours
  neighbours=s[(m*i):((m*i)+k)]
  g.append(neighbours)
G=np.vstack(g) 
f=G.astype(int)  #converting all type float datasets to type int 
b=[]
z=[]
for i in f[:,2]:
  b.append(y[i])  
for i in range(mt):
  res=max(b[k*i:k*(i+1)],key=b[k*i:k*(i+1)].count)  #getting maximum count of probable label
  z.append(res)
z_arr=np.array(z)
y_p=z_arr.reshape(mt,1)
y_s=yt[:mt].reshape(mt,1)
diff=y_p-y_s
q=0
for i in range(mt):
  if diff[i]==0:
    q+=1
acc=(q/mt)*100 #accuracy percentage of model
acc       
  

    

 
  
```
