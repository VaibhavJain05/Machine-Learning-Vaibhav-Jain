```python
train = train_set.to_numpy() #converting datasets from csv to numpy arrays
test = test_set.to_numpy()
```

```python
x= (train[:,1:])/255
xt= (test[:,1:])/255
xt
```

```python
y=train[:,0]
yt=test[:,0]
```

```python
def dist():  
  w=[]
  for i in range(1000):
    for j in range(1000):
      diff=xt[i,:]-x[j,:]
      dis=np.sqrt(np.sum(np.square(diff)))
      w.append((i,dis,j))
  return w 
s=sorted(dist())
k=9
g=[]
for i in range(1000):
  neighbours=s[(1000*i):((1000*i)+k)]
  g.append(neighbours)
G=np.vstack(g) 
f=G.astype(int)
m=[]
z=[]
for i in f[:,2]:
  m.append(y[i])  
for i in range(1000):
  res=max(m[k*i:k*(i+1)],key=m[k*i:k*(i+1)].count)
  z.append(res)
z_arr=np.array(z)
z_arr.reshape(1000,1)
y_s=yt[:1000]
y_p=y_s.reshape(1000,1)
diff=z_arr-y_s
q=0
for i in range(1000):
  if diff[i]==0:
    q+=1
acc=(q/1000)*100
acc       
  

    

 
  
```
